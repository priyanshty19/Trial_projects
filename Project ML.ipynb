```python
def add_age(cols):
    Age=cols[0];
    Pclass=cols[1];
    if pd.isnull(Age):
        return int(df[df["Pclass"]==Pclass]["Age"].mean())
    else:
        return Age
```

```python
df["Age"]=df[["Age","Pclass"]].apply(add_age,axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop("Cabin",inplace=True,axis=1)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
pd.get_dummies(df["Sex"])
```

```python
sex=pd.get_dummies(df["Sex"],drop_first=True)
```

```python
embarked=pd.get_dummies(df["Embarked"],drop_first=True)
pclass=pd.get_dummies(df["Pclass"],drop_first=True)
```

```python
df=pd.concat([df,pclass,sex,embarked],axis=1)
```

```python
df.head()
```

```python
# df.drop("Pclass",inplace=True,axis=1)
# df.drop("Embarkwed",inplace=True,axis=1)
df.drop("Embarked",inplace=True,axis=1)
df.head()
```

```python
df.drop(["Name","Ticket","Sex"],inplace=True,axis=1)
df.head()
```

```python
X=df.drop("Survived",axis=1)
Y=df["Survived"]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.25,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression(random_state=101)
logmodel.fit(X_train,Y_train)
```

```python
logmodel=LogisticRegression(max_iter=500,random_state=101)
logmodel.fit(X_train,Y_train)
```

```python
prediction=logmodel.predict(X_test)
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(Y_test,prediction))
```

```python
print(accuracy_score(Y_test,prediction))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,prediction)
```
